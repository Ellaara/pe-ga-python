```python
#filter base pay greater than 80K, only return EmployeeName and Job Title

salaries[salaries['BasePay'] > 80000][['EmployeeName','JobTitle']]
```

```python
#multiple conditions
#base pay greater than 80K and from San Fransisco Agency
salaries[(salaries['BasePay'] > 80000) & (salaries['Agency'] == 'San Francisco')] [['EmployeeName', 'JobTitle']]

#1. wrap each condition in round brackets
#2. we use '&' instead of the word 'and'
```
